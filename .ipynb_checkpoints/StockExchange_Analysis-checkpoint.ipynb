## Data

The data retrieved from Kaggle contained two dataframes: `indexinfo.csv` and `indexData.csv`.


`indexinfo.csv` contains information on each index referenced in the `indexData.csv` dataframe. The dataframe has four columns of information on the 14 unique indexes:
```python
[Region, Exchange, Index, Currency]
```

`indexData.csv` contains 112,457 rows of day by day information for the indexes broken down into 8 columns:  
```python
 [index_name, date, open, high, low, close, adj_close, volume]
```

The information in the dataframe is sourced from Yahoo Finance dating as far back as January 5th, 1965 and as most recently as June 3rd, 2021. Not every index starts at the same date, however each index ends at approximately the same time. The price of each stock is quoted in the national currency where each index is located.

````python
### Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
````

````python
### Importing CSV file
stock_info_df = pd.read_csv("indexInfo.csv")
stock_info_df.columns = ["region", "exchange", 'index_name', "currency"]
stock_df_original = pd.read_csv("indexData.csv")
df_columns_adjusted = stock_df_original
df_columns_adjusted.columns = ["index_name", "date", "open", "high", "low", "close", "adj_close", "volume"]
````

````python
### Creating cleaned and broken down dataframes


### Removing duplicates
df_duplicates_dropped = df_columns_adjusted.dropna()

### Merging info and data dataframes
df_merged = pd.merge(df_duplicates_dropped, stock_info_df, how="left", on="index_name")

### Creating individual dataframes that feature an aggregate of average adj_close price for each year + normalized
NYA_df = pd.DataFrame()
IXIC_df = pd.DataFrame()
HSI_df = pd.DataFrame()
SS_df = pd.DataFrame()
GSPTSE_df = pd.DataFrame()
SZ_df = pd.DataFrame()
NSEI_df = pd.DataFrame()
GDAXI_df = pd.DataFrame()
KS11_df = pd.DataFrame()
SSMI_df = pd.DataFrame()
TWII_df = pd.DataFrame()
JO_df = pd.DataFrame()
N225_df = pd.DataFrame()
N100_df = pd.DataFrame()
df_norm_adjclose = pd.DataFrame()

ind_df_list = [NYA_df, IXIC_df, HSI_df, SS_df, GSPTSE_df, SZ_df, NSEI_df,
           GDAXI_df, KS11_df, SSMI_df, TWII_df, JO_df, N225_df, N100_df]
ind_name_list = df_merged.index_name.unique().tolist()

for i in range(len(ind_df_list)):
    workable_df = df_merged[df_merged.index_name == ind_name_list[i]].reset_index()
    workable_df["year"] = workable_df.date.str[:4]
    workable_df = workable_df.groupby(["index_name","year"]).adj_close.mean().reset_index()
    
    adjclose_values = workable_df.adj_close.values
    adjclose_values = np.array(adjclose_values)
    adjclose_values = adjclose_values.reshape(len(adjclose_values),1)
    scaler = MinMaxScaler(feature_range=(0,1))
    adjclose_norm = scaler.fit_transform(adjclose_values)
    
    workable_df["norm"] = list(float(x) for x in adjclose_norm)
    ind_df_list[i] = workable_df
    
### Leaving space below to create new dataframes based on region, currency, or index
````

````python
def Feature_Frequency_Bar(data_f, colorbar, fignum, dropnum):
    sns.set_style("darkgrid")
    sns.set_context("talk")

    pct_index_df = df_merged[data_f].value_counts().reset_index()
    pct_index_df["percentage"] = pct_index_df.apply(lambda x: x[data_f]/sum(pct_index_df[data_f])*100, axis=1).round(decimals=2)
    pct_index_df.columns = [data_f, "counts", "percentage"]

    fig, ax = plt.subplots(figsize=(20,10))
    ax = sns.barplot(x=data_f, y="counts", data=pct_index_df, palette=colorbar)
    plt.xlabel(data_f.title(), labelpad=20)
    plt.ylabel("Count")
    plt.title("Frequency of Each {} in the Data".format(data_f.title()), fontsize=25, pad=10)

    i=0
    for p in ax.patches:
        if i < 3:
            ax.text(x = p.get_x() + p.get_width() / 2, 
                y = p.get_height() - dropnum,
                s = str(pct_index_df.percentage[i])+"%", 
                ha = 'center',
                color = 'white')
        else:
            ax.text(x = p.get_x() + p.get_width() / 2, 
                y = p.get_height() - dropnum,
                s = str(pct_index_df.percentage[i])+"%", 
                ha = 'center',
                color = 'black')
            
        i+=1
    
    plt.figtext(.01, .01, "Figure 1.{}".format(fignum), fontsize=15, fontstyle="italic")
    
    plt.show()
````

````python
Feature_Frequency_Bar("index_name", "Blues_r", 1, 750)
Feature_Frequency_Bar("region", "Oranges_r", 2, 1500)
Feature_Frequency_Bar("currency", "Greens_r", 3, 1500)
````
